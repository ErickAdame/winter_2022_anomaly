```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
file_input = "./Resources/nyc_jan_temps.csv"
nyc_df = pd.read_csv(file_input)
nyc_df
```

```python
nyc_df.columns
```

```python
nyc_df.dropna()
nyc_clean = nyc_df.rename(columns={"New York (Central Park)": "Year",
                       " New York": "Average Temperature",
                       " Average Temperature": "Anomaly"
                      })
```

```python
nyc_clean
```

```python
nyc_clean = nyc_clean.drop([0,1,2,3])
```

```python
nyc_clean
```

```python
del nyc_clean[" January"]
```

```python
nyc_clean["Year"] = nyc_clean["Year"].str[:-2]
```

```python
nyc_clean
```

```python
nyc_clean["Average Temperature"] = pd.to_numeric(nyc_clean["Average Temperature"])
nyc_clean["Year"] = pd.to_numeric(nyc_clean["Year"])
del nyc_clean["Anomaly"]
nyc_clean = nyc_clean[nyc_clean["Year"]>1960]
```

```python
nyc_clean.index = nyc_clean["Year"]
del nyc_clean["Year"]
```

```python
%matplotlib notebook
nyc_clean.plot(kind="line", title = "NYC: Average January Temperature", xlabel = "YEAR", ylabel = "DEGREES (F)", figsize=(7,4), legend=False)

plt.axhline(y = 33.7, color = 'r', linestyle = '--')
plt.savefig("./nyc_jan_temps.png", transparent=True)
```

```python
x_values = range(62)
y_values = nyc_clean["Average Temperature"]

slope = stats.linregress(x_values,y_values)
y_fit = slope[0]*x_values + slope[1]

plt.plot(x_values, y_fit, "-", color="black", alpha=0.25)
```
