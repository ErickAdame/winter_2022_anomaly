```python
import pandas as pd
import numpy as np
```

```python
file_input = "./Resources/nyc_jan_temps.csv"
nyc_df = pd.read_csv(file_input)
nyc_df
```

```python
nyc_df.dropna()
nyc_df.rename(columns={"New York (Central Park)": "Year",
                       " New York": "Average Temperature",
                       " Average Temperature": "Anomaly"
                      })
```

```python
clean_df = nyc_df.drop([0,1,2,3])
```

```python
clean_df.rename(columns={"New York (Central Park)": "Year",
                       " New York": "Average Temperature",
                       " Average Temperature": "Anomaly"
                      })
```

```python
del clean_df[" January"]
```

```python
clean_df
```

```python
years = pd.Series(range(1895,2023))
years
```

```python
pd.to_numeric(clean_df[" New York"])
jan_temps = clean_df[" New York"].astype(np.float64)
final_nyc_temps = pd.DataFrame(
    {
        "Average Temperature": jan_temps
    }
)
final_nyc_temps = final_nyc_temps.set_index(years)
final_nyc_temps
```
